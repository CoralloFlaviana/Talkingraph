```python
import pandas as pd
import simpledorff
```

```python
#folder_path = "/content/drive/MyDrive/Progetti Mike-Pier/Talkingraph/evaluation"
folder_path = "."
file_name = "talkingraph_eval_mike_pier.xlsx"
```

```python
def safe_krippendorff(df, class_col):
    unique_values = df[class_col].dropna().unique()
    if len(unique_values) <= 1:
        print(f"Non abbastanza variazione nei dati per calcolare Krippendorff's alpha per {class_col}")
        return 1
    try:
        return simpledorff.calculate_krippendorffs_alpha_for_df(
            df,
            experiment_col='unit_id',
            annotator_col='annotator_id',
            class_col=class_col
        )
    except ZeroDivisionError:
        print(f"Divisione per zero durante il calcolo di Krippendorff's alpha per {class_col} â€” expected_disagreement = 0")
        return 1
```

## Alpha di Krippendorffs llama

```python
# read pier annotations
df_pier = pd.read_excel(f'{folder_path}/{file_name}', sheet_name='llama-pier')

# mantieni solamente la colonna 2 e 3
df_pier = df_pier[['#INCORRECT', '#NOT_CHECKABLE', '#MISLEADING', '#OTHER']]

# ripristina l'indice
df_pier = df_pier.reset_index(drop=True)

# ottieni solo le prime 50 righe
df_pier = df_pier.iloc[:50]

# read pier annotations
df_michael = pd.read_excel(f'{folder_path}/{file_name}', sheet_name='llama-mike')

# mantieni solamente la colonna 2 e 3
df_michael = df_michael[['#INCORRECT', '#NOT_CHECKABLE', '#MISLEADING', '#OTHER']]

# ripristina l'indice
df_michael = df_michael.reset_index(drop=True)

# ottieni solo le prime 50 righe
df_michael = df_michael.iloc[:50]

# Unisco le annotazioni
unit_ids = list(range(len(df_pier)))
annotator1_id = 'P'
annotator2_id = 'M'

# Creo un DataFrame combinato
combined_data = []
for i in unit_ids:
    combined_data.append([i, annotator1_id, df_pier.loc[i, '#INCORRECT'], df_pier.loc[i, '#NOT_CHECKABLE'], df_pier.loc[i, '#MISLEADING'], df_pier.loc[i, '#OTHER']])
    combined_data.append([i, annotator2_id, df_michael.loc[i, '#INCORRECT'], df_michael.loc[i, '#NOT_CHECKABLE'], df_michael.loc[i, '#MISLEADING'], df_michael.loc[i, '#OTHER']])

combined_df = pd.DataFrame(combined_data, columns=['unit_id', 'annotator_id', '#INCORRECT', '#NOT_CHECKABLE', '#MISLEADING', '#OTHER'])

alpha_incorrect = safe_krippendorff(combined_df, '#INCORRECT')
alpha_not_checkable = safe_krippendorff(combined_df, '#NOT_CHECKABLE')
alpha_misleading = safe_krippendorff(combined_df, '#MISLEADING')
alpha_other = safe_krippendorff(combined_df, '#OTHER')

print(f"Krippendorff's alpha per llama incorrect: {alpha_incorrect}")
print(f"Krippendorff's alpha per llama not checkable: {alpha_not_checkable}")
print(f"Krippendorff's alpha per llama misleading: {alpha_misleading}")
print(f"Krippendorff's alpha per llama other: {alpha_other}")

total = alpha_incorrect + alpha_not_checkable + alpha_misleading + alpha_other
```
